```python
import pandas as pd
music_rank = pd.read_csv(r'C:\Users\AdminA\Desktop\Blossom files\Assign1\billboard.csv')
music_rank
```

```python
music_rank_melt = pd.melt(music_rank, id_vars=['artiste','track','genre','date.entered','date.peaked'])
music_rank_melt
```

```python
music_rank_melt = music_rank_melt.rename(columns={'variable':'week','value':'rank'})
music_rank_melt
```

```python
music_rank_melt.dtypes
```

```python
music_rank_melt['week'] = music_rank_melt['week'].str.extract('(\d+)').astype(int)
music_rank_melt
```

```python
music_rank_melt.dtypes
```

```python
music_rank_melt['date.entered'] = pd.to_datetime(music_rank_melt['date.entered'])
music_rank_melt['date.peaked'] = pd.to_datetime(music_rank_melt['date.peaked'])
```

```python
music_rank_melt['entered_to_peak'] = music_rank_melt['date.peaked'] - music_rank_melt['date.entered']
music_rank_melt
```

```python
music_rank_melt['entered_to_peak'] = music_rank_melt['entered_to_peak'].dt.days
```
